# Simple Moving Average

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
%matplotlib inline
```

```python
data = pd.read_csv('f_p1.csv',parse_dates=True)
data.head()
```

```python
del data['Unnamed: 0']
```

```python
data.head()
data.info()
```

# Data Slicing 0 to 5000


```python
df = data['UC'].iloc[0:197].to_frame()
```

### Simple Moving Average: 30

```python
df['SMA30'] = df['UC'].rolling(30).mean()
# df.dropna(inplace=True)
df
```

```python
df[['UC', 'SMA30']].plot(label='UC',
                                  figsize=(16, 10))
```
